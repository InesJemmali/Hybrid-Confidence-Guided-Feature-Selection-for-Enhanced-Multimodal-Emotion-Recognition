```python
import os
import torch
import numpy as np
import cv2
from PIL import Image
from facenet_pytorch import MTCNN

# Initialize MTCNN with landmark output
device = 'cuda' if torch.cuda.is_available() else 'cpu'
mtcnn = MTCNN(keep_all=False, device=device)

# Paths
combined_txt = r"D:\Nouveau dossier\data\data\mosei\two_emotion_folders2.txt"
data_root = "D:\\Nouveau dossier\\data\\data\\MOSEI_RAW_PROCESSED"

# Parameters
output_size = 128
margin_ratio = 0.3

# Load sample names
with open(combined_txt, 'r') as f:
    sample_names = [line.strip() for line in f.readlines()]

# Track samples with <4 faces
few_faces_samples = []

# Align face using eye landmarks
def align_face(img, landmarks, output_size=128, margin_ratio=0.3):
    # Landmarks: [left_eye, right_eye, nose, mouth_left, mouth_right]
    left_eye, right_eye = landmarks[0], landmarks[1]

    # Compute angle between eyes
    dx = right_eye[0] - left_eye[0]
    dy = right_eye[1] - left_eye[1]
    angle = np.degrees(np.arctan2(dy, dx))

    # Center between eyes
    eye_center = ((left_eye[0] + right_eye[0]) / 2, (left_eye[1] + right_eye[1]) / 2)

    # Convert PIL to OpenCV
    img_cv = np.array(img)
    img_cv = cv2.cvtColor(img_cv, cv2.COLOR_RGB2BGR)

    # Rotate image around eye center
    rot_mat = cv2.getRotationMatrix2D(eye_center, angle, scale=1.0)
    rotated = cv2.warpAffine(img_cv, rot_mat, (img_cv.shape[1], img_cv.shape[0]))

    # Apply same rotation to landmarks
    ones = np.ones((landmarks.shape[0], 1))
    landmarks_hom = np.hstack([landmarks, ones])
    rotated_landmarks = rot_mat @ landmarks_hom.T
    rotated_landmarks = rotated_landmarks.T

    # Bounding box with margin
    x1 = np.min(rotated_landmarks[:, 0])
    y1 = np.min(rotated_landmarks[:, 1])
    x2 = np.max(rotated_landmarks[:, 0])
    y2 = np.max(rotated_landmarks[:, 1])

    cx, cy = (x1 + x2) / 2, (y1 + y2) / 2
    w, h = x2 - x1, y2 - y1
    margin = margin_ratio * max(w, h)

    crop_x1 = int(max(cx - output_size / 2 - margin, 0))
    crop_y1 = int(max(cy - output_size / 2 - margin, 0))
    crop_x2 = int(min(cx + output_size / 2 + margin, rotated.shape[1]))
    crop_y2 = int(min(cy + output_size / 2 + margin, rotated.shape[0]))

    face_crop = rotated[crop_y1:crop_y2, crop_x1:crop_x2]
    face_crop = cv2.resize(face_crop, (output_size, output_size))

    # Convert back to PIL
    face_crop_pil = Image.fromarray(cv2.cvtColor(face_crop, cv2.COLOR_BGR2RGB))
    return face_crop_pil

# Process each sample
for sample in sample_names:
    sample_dir = os.path.join(data_root, sample)
    if not os.path.isdir(sample_dir):
        print(f"[Skip] Not a folder: {sample_dir}")
        continue

    image_files = sorted([f for f in os.listdir(sample_dir) if f.lower().endswith(('.jpg', '.png'))])
    faces_dir = os.path.join(sample_dir, "faces")
    os.makedirs(faces_dir, exist_ok=True)

    face_count = 0

    for img_file in image_files:
        img_path = os.path.join(sample_dir, img_file)
        face_path = os.path.join(faces_dir, img_file)

        try:
            img = Image.open(img_path).convert('RGB')
            boxes, probs, landmarks = mtcnn.detect(img, landmarks=True)

            if boxes is not None and landmarks is not None and probs[0] > 0.90:
                aligned = align_face(img, landmarks[0])  # align first face
                aligned.save(face_path)
                face_count += 1
            else:
                print(f"[No face] {img_file} in {sample}")
        except Exception as e:
            print(f"[Error] {img_file} in {sample}: {e}")

    # Check how many faces were extracted
    if face_count < 4:
        few_faces_samples.append(sample)

# Log samples with few faces
print("\n[Summary] Samples with <4 face frames:")
for s in few_faces_samples:
    print(f"- {s}")

# Optionally save to file
with open("samples_with_few_faces.txt", "w") as f:
    for s in few_faces_samples:
        f.write(s + "\n")
```

```python
import os
import torchaudio
import numpy as np

# Paths
name_file = r"D:\Nouveau dossier\data\data\MOSEI_SPLIT\file.txt"
data_base = r"D:\Nouveau dossier\data\data\MOSEI_RAW_PROCESSED"

# Target sample rate
target_sr = 8000

# Storage
audio_lengths = []
frame_counts = []

# Read folder names
with open(name_file, 'r') as f:
    folder_names = [line.strip() for line in f if line.strip()]

# Iterate through each sample
for folder in folder_names:
    folder_path = os.path.join(data_base, folder)
    audio_path = os.path.join(folder_path, "audio.wav")
    faces_path = os.path.join(folder_path, "faces")

    # --- Process audio ---
    if os.path.isfile(audio_path):
        try:
            waveform, sr = torchaudio.load(audio_path)
            if sr != target_sr:
                resampler = torchaudio.transforms.Resample(orig_freq=sr, new_freq=target_sr)
                waveform = resampler(waveform)
            audio_lengths.append(waveform.size(1))
        except Exception as e:
            print(f"[Error] Failed to process audio {audio_path}: {e}")
    else:
        print(f"[Warning] Missing audio: {audio_path}")

    # --- Process faces ---
    if os.path.isdir(faces_path):
        image_files = [f for f in os.listdir(faces_path) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
        frame_counts.append(len(image_files))
    else:
        print(f"[Warning] Missing faces folder: {faces_path}")

# --- Statistics helper ---
def print_stats(label, data):
    if not data:
        print(f"\n❌ No data found for {label}")
        return
    arr = np.array(data)
    print(f"\n📊 {label} Statistics:")
    print(f"   ➤ Min    : {arr.min()}")
    print(f"   ➤ Max    : {arr.max()}")
    print(f"   ➤ Mean   : {arr.mean():.2f}")
    print(f"   ➤ Median : {np.median(arr):.2f}")
    print(f"   ➤ Std    : {arr.std():.2f}")
    print(f"   ➤ 95th % : {np.percentile(arr, 95):.2f}")

# --- Final results ---
print_stats("Resampled Audio Lengths (at 8kHz, in samples)", audio_lengths)
print_stats("Number of Face Images per Sample", frame_counts)


```

```python
import os
import pickle
import random
from collections import defaultdict

# Paths
combined_path = r"D:\Nouveau dossier\data\data\mosei\finalcombined_filtered.txt"
pkl_path = r"D:\Nouveau dossier\data\data\mosei\filtered_data_named_labels.pkl"
output_dir = r"D:\Nouveau dossier\data\data\mosei"
os.makedirs(output_dir, exist_ok=True)

# Load list of sample folder names
with open(combined_path, 'r') as f:
    folder_names = [line.strip() for line in f if line.strip()]

# Load emotion labels
with open(pkl_path, 'rb') as f:
    label_dict = pickle.load(f)

# Group folders by primary emotion
emotion_to_folders = defaultdict(list)
unlabeled = []

for folder in folder_names:
    if folder in label_dict:
        label_vector = label_dict[folder]['label']
        if 1 in label_vector:
            main_emotion = label_vector.index(1)  # pick first 1 as emotion index
            emotion_to_folders[main_emotion].append(folder)
        else:
            unlabeled.append(folder)
            print(f"[Warning] No active label for {folder}")
    else:
        unlabeled.append(folder)
        print(f"[Warning] Missing from label dict: {folder}")

# Stratified 85/5/10 split per emotion
train_set, val_set, test_set = set(), set(), set()
random.seed(42)

for emotion, folders in emotion_to_folders.items():
    random.shuffle(folders)
    n = len(folders)
    n_train = int(0.85 * n)
    n_val = int(0.05 * n)
    n_test = n - n_train - n_val  # all remaining

    train_set.update(folders[:n_train])
    val_set.update(folders[n_train:n_train + n_val])
    test_set.update(folders[n_train + n_val:])

# Ensure disjoint sets
assert train_set.isdisjoint(val_set)
assert train_set.isdisjoint(test_set)
assert val_set.isdisjoint(test_set)

# Save splits
def save_split(split_set, name):
    path = os.path.join(output_dir, f"{name}_split.txt")
    with open(path, 'w') as f:
        for item in sorted(split_set):
            f.write(item + '\n')
    print(f"✅ Saved {len(split_set)} to {path}")

save_split(train_set, "train")
save_split(val_set, "valid")
save_split(test_set, "test")

# Summary
print(f"\n📊 Processed labeled folders: {len(train_set) + len(val_set) + len(test_set)}")
print(f"❌ Unlabeled or missing: {len(unlabeled)}")

```

```python
import os
import torch
import numpy as np
import cv2
from PIL import Image
from facenet_pytorch import MTCNN
import shutil

# Device setup
device = 'cuda' if torch.cuda.is_available() else 'cpu'
mtcnn = MTCNN(keep_all=False, device=device)

# Paths
data_root = r"D:\Nouveau dossier\data\data\MOSEI_RAW_PROCESSED"

# You must already have the list `found_folders`
# Example: found_folders = ['abc_1', 'xyz_3', ...]

# Parameters
output_size = 128
margin_ratio = 0.3

# Track samples with <4 faces
few_faces_samples = []

# Align face using eye landmarks
def align_face(img, landmarks, output_size=128, margin_ratio=0.3):
    left_eye, right_eye = landmarks[0], landmarks[1]
    dx = right_eye[0] - left_eye[0]
    dy = right_eye[1] - left_eye[1]
    angle = np.degrees(np.arctan2(dy, dx))
    eye_center = ((left_eye[0] + right_eye[0]) / 2, (left_eye[1] + right_eye[1]) / 2)
    img_cv = np.array(img)
    img_cv = cv2.cvtColor(img_cv, cv2.COLOR_RGB2BGR)
    rot_mat = cv2.getRotationMatrix2D(eye_center, angle, scale=1.0)
    rotated = cv2.warpAffine(img_cv, rot_mat, (img_cv.shape[1], img_cv.shape[0]))

    ones = np.ones((landmarks.shape[0], 1))
    landmarks_hom = np.hstack([landmarks, ones])
    rotated_landmarks = rot_mat @ landmarks_hom.T
    rotated_landmarks = rotated_landmarks.T

    x1 = np.min(rotated_landmarks[:, 0])
    y1 = np.min(rotated_landmarks[:, 1])
    x2 = np.max(rotated_landmarks[:, 0])
    y2 = np.max(rotated_landmarks[:, 1])
    cx, cy = (x1 + x2) / 2, (y1 + y2) / 2
    w, h = x2 - x1, y2 - y1
    margin = margin_ratio * max(w, h)

    crop_x1 = int(max(cx - output_size / 2 - margin, 0))
    crop_y1 = int(max(cy - output_size / 2 - margin, 0))
    crop_x2 = int(min(cx + output_size / 2 + margin, rotated.shape[1]))
    crop_y2 = int(min(cy + output_size / 2 + margin, rotated.shape[0]))

    face_crop = rotated[crop_y1:crop_y2, crop_x1:crop_x2]
    face_crop = cv2.resize(face_crop, (output_size, output_size))
    face_crop_pil = Image.fromarray(cv2.cvtColor(face_crop, cv2.COLOR_BGR2RGB))
    return face_crop_pil

# Process each sample in found_folders
for sample in found_folders:
    sample_dir = os.path.join(data_root, sample)
    if not os.path.isdir(sample_dir):
        print(f"[Skip] Not a folder: {sample_dir}")
        continue

    # Overwrite 'faces/' folder if it exists
    faces_dir = os.path.join(sample_dir, "faces")
    if os.path.exists(faces_dir):
        shutil.rmtree(faces_dir)
    os.makedirs(faces_dir, exist_ok=True)

    # List image frames
    image_files = sorted([f for f in os.listdir(sample_dir)
                          if f.lower().endswith(('.jpg', '.png')) and os.path.isfile(os.path.join(sample_dir, f))])
    face_count = 0

    for img_file in image_files:
        img_path = os.path.join(sample_dir, img_file)
        face_path = os.path.join(faces_dir, img_file)

        try:
            img = Image.open(img_path).convert('RGB')
            boxes, probs, landmarks = mtcnn.detect(img, landmarks=True)

            if boxes is not None and landmarks is not None and probs[0] > 0.90:
                aligned = align_face(img, landmarks[0])
                aligned.save(face_path)
                face_count += 1
            else:
                print(f"[No face] {img_file} in {sample}")
        except Exception as e:
            print(f"[Error] {img_file} in {sample}: {e}")

    if face_count < 4:
        few_faces_samples.append(sample)

# Log and save the result
print("\n[Summary] Samples with <4 aligned face frames:")
for s in few_faces_samples:
    print(f"- {s}")

with open("samples_with_few_faces.txt", "w") as f:
    for s in few_faces_samples:
        f.write(s + "\n")
```

```python
import os
import pickle
import random

# Set seed for reproducibility
random.seed(42)

# Path to MOSEI label file
label_file = r"D:\Nouveau dossier\data\data\MOSEI_RAW_PROCESSED\filtered_data_named_labels.pkl"

# Paths to split files
train_path = r"D:\Nouveau dossier\data\data\MOSEI_SPLIT\train_split.txt"
valid_path = r"D:\Nouveau dossier\data\data\MOSEI_SPLIT\valid_split.txt"
test_path  = r"D:\Nouveau dossier\data\data\MOSEI_SPLIT\test_split.txt"

# Your existing found_folders list must be in memory
# found_folders = [...]

# Load labels
with open(label_file, 'rb') as f:
    data = pickle.load(f)

# Load existing splits
def load_split(path):
    if not os.path.exists(path):
        return set()
    with open(path, 'r') as f:
        return set(line.strip() for line in f.readlines())

train_set = load_split(train_path)
valid_set = load_split(valid_path)
test_set  = load_split(test_path)
already_used = train_set.union(valid_set).union(test_set)

# Emotion categories
emotions = ['anger', 'disgust', 'fear', 'happy', 'sad', 'surprise']

# Only target anger
emotion_targets = {
    'train': {'anger': 100},
    'valid': {'anger': 20},
    'test':  {'anger': 30}
}

# Bucket one-emotion samples for anger only
anger_bucket = []
for name in found_folders:
    if name in data and sum(data[name]['label']) == 1:
        label_idx = data[name]['label'].index(1)
        emotion = emotions[label_idx]
        if emotion == 'anger' and name not in already_used:
            anger_bucket.append(name)

# Shuffle anger samples
random.shuffle(anger_bucket)

# Allocate helper
def allocate_split(bucket, count):
    selected = bucket[:count]
    del bucket[:count]
    return selected

# Split allocation
split_names = {'train': [], 'valid': [], 'test': []}
split_names['train'].extend(allocate_split(anger_bucket, emotion_targets['train']['anger']))
split_names['valid'].extend(allocate_split(anger_bucket, emotion_targets['valid']['anger']))
split_names['test'].extend(allocate_split(anger_bucket, emotion_targets['test']['anger']))

# Save with deduplication
def append_to_file(path, names, existing_set):
    with open(path, 'a') as f:
        for name in names:
            if name not in existing_set:
                f.write(name + '\n')
                existing_set.add(name)

append_to_file(train_path, split_names['train'], train_set)
append_to_file(valid_path, split_names['valid'], valid_set)
append_to_file(test_path,  split_names['test'],  test_set)

# Report
print(f"Added to train: {len(split_names['train'])}")
print(f"Added to valid: {len(split_names['valid'])}")
print(f"Added to test : {len(split_names['test'])}")

```
